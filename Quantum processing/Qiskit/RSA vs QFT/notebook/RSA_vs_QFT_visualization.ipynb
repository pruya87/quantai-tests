```python
# Full Interactive Shor's Algorithm Simulation for Google Colab or Jupyter

import numpy as np
import matplotlib.pyplot as plt
from math import gcd
from fractions import Fraction
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

def shor_simulator(N, g, x_measured, Q=32):
    clear_output(wait=True)
    display(Markdown(f"## 🔐 Shor's Algorithm Simulation"))
    display(Markdown(f"**Modulus (N):** {N}  \n**Base (g):** {g}  \n**Simulated QFT Output (x):** {x_measured}  \n**Register Size (Q):** {Q}"))

    # Step 1: Continued fraction to get period
    frac = Fraction(x_measured, Q).limit_denominator(N)
    r = frac.denominator
    display(Markdown(f"🔁 Continued Fraction: $\\frac{{{x_measured}}}{{{Q}}} = {frac}$ → **Estimated period $r = {r}$**"))

    if r % 2 != 0:
        display(Markdown("⚠️ Period is **not even** — retry with a different `x` or `g`."))
        return

    # Step 2: Compute g^(r/2) mod N
    c = pow(g, r // 2, N)
    display(Markdown(f"🔐 $c = g^{{r/2}} \\mod N = {g}^{{{r//2}}} \\mod {N} = {c}$"))

    if c == 1 or c == N - 1:
        display(Markdown("⚠️ Trivial root: $g^{r/2} \\equiv \\pm1 \\mod N$ — retry with a different base `g`."))
        return

    # Step 3: Compute GCDs
    f1 = gcd(c + 1, N)
    f2 = gcd(c - 1, N)

    display(Markdown(f"🧮 $\\gcd(c + 1, N) = \\gcd({c + 1}, {N}) = {f1}$  \n🧮 $\\gcd(c - 1, N) = \\gcd({c - 1}, {N}) = {f2}$"))

    if f1 in [1, N] and f2 in [1, N]:
        display(Markdown("❌ No nontrivial factors found — try different values."))
    else:
        display(Markdown(f"✅ **Success!** Nontrivial factors of **{N}** are: **{f1}** and **{f2}**"))

    # Step 4: Visualize probability distribution for correct vs wrong r
    x_vals = np.arange(Q)
    prob_true = np.sin(np.pi * r * x_vals / Q)**2
    prob_true /= prob_true.sum()

    r_wrong = r + 1 if r + 1 < Q else r - 1
    prob_wrong = np.sin(np.pi * r_wrong * x_vals / Q)**2
    prob_wrong /= prob_wrong.sum()

    fig, axs = plt.subplots(2, 1, figsize=(12, 5), sharex=True)
    axs[0].bar(x_vals, prob_true, color="green")
    axs[0].set_title(f"✅ Simulated Probability (Correct r = {r})")
    axs[1].bar(x_vals, prob_wrong, color="red")
    axs[1].set_title(f"❌ Simulated Probability (Wrong r = {r_wrong})")
    plt.tight_layout()
    plt.show()


# Set up the interactive widget
widgets.interact(
    shor_simulator,
    N=widgets.Dropdown(options=[15, 21, 35, 77], value=15, description="Modulus N:"),
    g=widgets.IntSlider(min=2, max=10, step=1, value=2, description="Base g:"),
    x_measured=widgets.IntSlider(min=1, max=31, step=1, value=8, description="QFT x:"),
);
```
